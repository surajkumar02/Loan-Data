```python
#### dependent approaches for calculation
## new features provide additional information that could potentially improve the performance

# Create new features
loan_data['credit_utilization'] = loan_data['revol_bal'] / loan_data['total_acc']
loan_data['income_to_loan_ratio'] = loan_data['annual_inc'] / loan_data['loan_amnt']
loan_data['emp_length_years'] = loan_data['emp_length'].str.extract('(\d+)').astype(float)
loan_data['fico_score_range'] = loan_data['fico_range_high'] - loan_data['fico_range_low']

# Print the head of the dataframe to verify the new features
loan_data.head()
```

```python
# Define the features and target
loan_data_encoded =loan_data.copy()
# loan_data_encoded['loan_status']
loan_data_encoded['loan_status_Paid'] =loan_data_encoded['loan_status'].replace('Paid', float(1))
loan_data_encoded['loan_status_Paid'] =loan_data_encoded['loan_status_Paid'].replace('Defaulted', float(0))
loan_data_encoded['loan_status_Defaulted'] =loan_data_encoded['loan_status'].replace('Defaulted', float(1))
loan_data_encoded['loan_status_Defaulted'] =loan_data_encoded['loan_status_Defaulted'].replace('Paid', float(0))

# loan_data_encoded.columns

X = loan_data_encoded[['loan_status_Paid', 'loan_status_Defaulted']]
y = loan_data_encoded['loan_status_Paid']

# X, y = make_classification(n_classes=2, class_sep=2,
# weights=[0.1, 0.9], n_informative=3, n_redundant=1, flip_y=0,
# n_features=20, n_clusters_per_class=1, n_samples=2000, random_state=10)

# Apply SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

# Check the distribution of the target after SMOTE
pd.Series(y_res).value_counts()
```

```python
# Define the model
model = LogisticRegression(random_state=42)


# Define the parameters grid
params_grid = {'C': [0.10, 0.20, 0.30, 1, 10, 20, 40, 50, 75, 100]}

# Apply grid search with cross-validation
grid_search = GridSearchCV(model, params_grid, cv=5, scoring='accuracy')
grid_search.fit(X_res, y_res)

# Print the best parameters and the best score
grid_search.best_params_, grid_search.best_score_
```
